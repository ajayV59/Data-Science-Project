```python
# applying regex to remove rating in the text

import re

df['society'] = df['society'].apply(lambda x: re.sub(r'(\d+)?(\.\d+)?\s?\★','',str(x)).strip()).str.lower()
```

```python
#nan values count is 619 -> converting nan to independent

df['society'] = df['society'].str.replace('nan','independent')
```

```python
df = df[df['price'] != 'Price on Request']  #removing data having price on request
```

```python
df.head(3)
```

```python
# converting price column to float ain terms of 'crore' value

def price_convert_numeric(x):
        numeric_part1 = x.replace('Lac', '').strip()
        numeric_part2 = x.replace('Crore', '').strip()
        if 'Lac' in x:
            return round(float(numeric_part1)/100,2)
        else:
            return round(float(numeric_part2),2)
```

```python
x = df[pd.isna(df['price'])].index
```

```python
x
```

```python
df.drop(x,inplace=True) # dropping missing values from 'price' column
```

```python
df.info()
```

```python
df['price'] = df['price'].apply(price_convert_numeric)
```
